```python
import osmnx as ox
import pandas as pd
import networkx as nx
import warnings
warnings.filterwarnings('ignore')
import numpy as np
```

```python
G = ox.graph_from_point(center_point=(35.88050, 139.65564), network_type='all', dist=1000)
```

```python
# hardware result
# [node1, node2, count]
hw_result = []
with open('./CREST_DEMO.txt') as f:
    for line in f:
        line = line.rstrip()
        l = [int(x.strip()) for x in line.split(',')]
        hw_result.append(l)
```

```python
# ハードウェアで公開される経路
def hw_disclosure_count(k):
    cnt = 0
    hw_path = []
    for i in range(len(hw_result)):
        if hw_result[i][2] >= k and hw_result[i][0] != hw_result[i][1] and hw_result[i][0] in G and hw_result[i][1] :
            cnt+=1
            hw_path.append([hw_result[i][0], hw_result[i][1]])
    print(cnt)
    return hw_path
```

```python
# # path length check
def path_check(seq, graph):
    new_seq = []
    for index, i in enumerate(seq):
        try:
            # 最初の方向のパスを試みる
            path = ox.shortest_path(graph, i[0], i[1])
            if len(path) != 2:
                # 逆方向のパスも試みる
                reverse_path = ox.shortest_path(graph, i[1], i[0])
                if len(reverse_path) == 2:
                    new_seq.append(reverse_path)
            else:
                new_seq.append(path)
        except (nx.NodeNotFound, nx.NetworkXNoPath):
            pass
    return new_seq
```

```python
hw_k1_paths = path_check(hw_disclosure_count(1), G)
hw_k2_paths = path_check(hw_disclosure_count(2), G)
hw_k4_paths = path_check(hw_disclosure_count(4), G)
hw_k8_paths = path_check(hw_disclosure_count(8), G)
hw_k12_paths = path_check(hw_disclosure_count(12), G)
hw_k16_paths = path_check(hw_disclosure_count(16), G)
hw_k20_paths = path_check(hw_disclosure_count(20), G)
hw_k24_paths = path_check(hw_disclosure_count(24), G)
hw_k28_paths = path_check(hw_disclosure_count(28), G)
hw_k32_paths = path_check(hw_disclosure_count(32), G)
hw_k64_paths = path_check(hw_disclosure_count(64), G)
hw_k128_paths = path_check(hw_disclosure_count(128), G)
hw_k256_paths = path_check(hw_disclosure_count(256), G)
print("number of hw segment: ")
```

```python
# hardware segment(k=1)
print("number of hw segment: ", len(hw_k1_paths))
fig, ax = ox.plot_graph_routes(G, hw_k1_paths, node_size = 0, orig_dest_size = 0, route_linewidth = 0.5, route_alpha = 1)
```
